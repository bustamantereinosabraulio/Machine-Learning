## Notación

Aquí tienes un resumen de la notación que encontrarás en este laboratorio:

| General | Notación | Descripción | Python (si aplica) |
|:--------|:---------|:-------------|:-------------------|
| $a$ | escalar, sin negrita | |

| $\mathbf{x}$ | Valores de las características de un ejemplo de entrenamiento (en este lab: tamaño en miles de pies cuadrados) | `x_train` |

## Herramientas

En este laboratorio harás uso de:

- **NumPy**, una librería popular para computación científica.  
- **Matplotlib**, una librería popular para graficar datos.  


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# --- 1. Estilo por defecto ---
plt.figure(figsize=(6,4))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Estilo por defecto (Matplotlib)")
plt.legend()
plt.show()

# --- 2. Estilo ggplot ---
plt.style.use("ggplot")
plt.figure(figsize=(6,4))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Estilo ggplot (Matplotlib)")
plt.legend()
plt.show()

# --- 3. Estilo fivethirtyeight ---
plt.style.use("fivethirtyeight")
plt.figure(figsize=(6,4))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Estilo fivethirtyeight (Matplotlib)")
plt.legend()
plt.show()

# --- 4. Usando Seaborn ---
sns.set_theme(style="darkgrid")  # estilo de Seaborn
plt.figure(figsize=(6,4))
sns.lineplot(x=x, y=y1, label="sin(x)")
sns.lineplot(x=x, y=y2, label="cos(x)")
plt.title("Estilo con Seaborn")
plt.legend()
plt.show()
```
